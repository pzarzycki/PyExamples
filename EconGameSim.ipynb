```
# Pawel Zarzycki, 2020, SWPS University / PJATK
# Based on:
# Boghosian B. M. (2019). "The Inescapable Casino" in Scientific American 321, 5, 70-77 (November 2019) doi:10.1038/scientificamerican1119-70

import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from random import seed
from random import randint
import seaborn as sns

population = 10000
iterations = 10000000
iter_plot =  100000
init_v = 10.0
win_rate = 0.2
loss_rate = 0.1 # 0.17

#init
seed(1)
data = np.array([population * [init_v]])

for nn in range(1, iterations):
    #data = np.vstack ([data, data[n-1]])
    n = 0
    a = randint(0, population-1)    # a losses
    b = randint(0, population-1)    # b wins
    if (a == b):
        continue
    if (data[n,a] < data[n,b]):
        amount = data[n,a] * loss_rate
    else:
        amount = data[n,b] * win_rate
    data[n,a]-=amount
    data[n,b]+=amount

    if (nn == 1 or nn % iter_plot == 0):
        #plt.hist(data[n,:])
        print("Iteration: %d" % nn)
        sns.distplot(data[n], hist=False, kde=True, color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
        plt.show()
```
